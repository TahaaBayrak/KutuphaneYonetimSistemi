```python
from abc import ABC, abstractmethod

# --- SOYUT SINIFLAR ---
class Kaynak(ABC):
    def __init__(self, baslik, kayit_no):
        self._baslik = baslik
        self._kayit_no = kayit_no

    @property
    def baslik(self):
        return self._baslik

    @baslik.setter
    def baslik(self, deger):
        self._baslik = deger

    @property
    def kayit_no(self):
        return self._kayit_no

    @kayit_no.setter
    def kayit_no(self, deger):
        self._kayit_no = deger

class Islem(ABC):
    @abstractmethod
    def ekle(self):
        pass

    @abstractmethod
    def sil(self):
        pass

    @abstractmethod
    def guncelle(self):
        pass

    @abstractmethod
    def listele(self):
        pass


# --- ALT SINIFLAR (KAYNAK) ---
class Kitap(Kaynak):
    def __init__(self, baslik, kayit_no, yazar, sayfa_sayisi):
        super().__init__(baslik, kayit_no)
        self._yazar = yazar
        self._sayfa_sayisi = sayfa_sayisi

    @property
    def yazar(self):
        return self._yazar

    @yazar.setter
    def yazar(self, deger):
        self._yazar = deger

    @property
    def sayfa_sayisi(self):
        return self._sayfa_sayisi

    @sayfa_sayisi.setter
    def sayfa_sayisi(self, deger):
        self._sayfa_sayisi = deger

    def __str__(self):
        return f"Kayıt No: {self.kayit_no} | Kitap: {self.baslik} | Yazar: {self.yazar} | Sayfa: {self.sayfa_sayisi}"


class Dergi(Kaynak):
    def __init__(self, baslik, kayit_no, yayin_donemi, sayi_no):
        super().__init__(baslik, kayit_no)
        self._yayin_donemi = yayin_donemi
        self._sayi_no = sayi_no

    @property
    def yayin_donemi(self):
        return self._yayin_donemi

    @yayin_donemi.setter
    def yayin_donemi(self, deger):
        self._yayin_donemi = deger

    @property
    def sayi_no(self):
        return self._sayi_no

    @sayi_no.setter
    def sayi_no(self, deger):
        self._sayi_no = deger

    def __str__(self):
        return f"Kayıt No: {self.kayit_no} | Dergi: {self.baslik} | Dönem: {self.yayin_donemi} | Sayı: {self.sayi_no}"


# --- ALT SINIFLAR (İŞLEM) ---
class KitapIslem(Islem):
    def __init__(self):
        self._kitaplar = []

    def kitap_sayisi(self):
        return len(self._kitaplar)

    def _kayit_no_var_mi(self, kayit_no):
        for kitap in self._kitaplar:
            if kitap.kayit_no == kayit_no:
                return True
        return False

    def ekle(self):
        kayit_no = input("Kitabın kayıt numarasını girin: ")

        if self._kayit_no_var_mi(kayit_no):
            print("Hata: Bu kayıt numarası zaten var!")
            return

        baslik = input("Kitabın başlığını girin: ")
        yazar = input("Kitabın yazarını girin: ")
        sayfa = input("Kitabın sayfa sayısını girin: ")

        yeni_kitap = Kitap(baslik, kayit_no, yazar, sayfa)
        self._kitaplar.append(yeni_kitap)
        print("Kitap başarıyla eklendi.")
        print(f"Toplam Kitap Sayısı: {self.kitap_sayisi()}")

    def sil(self):
        silinecek_no = input("Silinecek kitabın kayıt numarasını girin: ")
        for kitap in self._kitaplar:
            if kitap.kayit_no == silinecek_no:
                self._kitaplar.remove(kitap)
                print("Kitap başarıyla silindi.")
                return
        print("Kayıt bulunamadı.")

    def guncelle(self):
        guncellenecek_no = input("Güncellenecek kitabın kayıt numarasını girin: ")
        for kitap in self._kitaplar:
            if kitap.kayit_no == guncellenecek_no:
                yeni_baslik = input("Yeni başlık (değişmeyecekse boş bırakın): ")
                yeni_yazar = input("Yeni yazar (değişmeyecekse boş bırakın): ")
                yeni_sayfa = input("Yeni sayfa sayısı (değişmeyecekse boş bırakın): ")

                if yeni_baslik != "":
                    kitap.baslik = yeni_baslik
                if yeni_yazar != "":
                    kitap.yazar = yeni_yazar
                if yeni_sayfa != "":
                    kitap.sayfa_sayisi = yeni_sayfa

                print("Kitap güncellendi.")
                return
        print("Kayıt bulunamadı.")

    def listele(self):
        if self.kitap_sayisi() == 0:
            print("Kayıt bulunamadı.")
        else:
            print("\n--- KİTAPLAR LİSTESİ ---")
            for kitap in self._kitaplar:
                print(kitap)


class DergiIslem(Islem):
    def __init__(self):
        self._dergiler = []

    def dergi_sayisi(self):
        return len(self._dergiler)

    def _kayit_no_var_mi(self, kayit_no):
        for dergi in self._dergiler:
            if dergi.kayit_no == kayit_no:
                return True
        return False

    def ekle(self):
        kayit_no = input("Derginin kayıt numarasını girin: ")

        if self._kayit_no_var_mi(kayit_no):
            print("Hata: Bu kayıt numarası zaten var!")
            return

        baslik = input("Derginin başlığını girin: ")
        yayin_donemi = input("Yayın dönemini girin (Aylık/Haftalık vb.): ")
        sayi_no = input("Sayı numarasını girin: ")

        yeni_dergi = Dergi(baslik, kayit_no, yayin_donemi, sayi_no)
        self._dergiler.append(yeni_dergi)
        print("Dergi başarıyla eklendi.")
        print(f"Toplam Dergi Sayısı: {self.dergi_sayisi()}")

    def sil(self):
        silinecek_no = input("Silinecek derginin kayıt numarasını girin: ")
        for dergi in self._dergiler:
            if dergi.kayit_no == silinecek_no:
                self._dergiler.remove(dergi)
                print("Dergi başarıyla silindi.")
                return
        print("Kayıt bulunamadı.")

    def guncelle(self):
        guncellenecek_no = input("Güncellenecek derginin kayıt numarasını girin: ")
        for dergi in self._dergiler:
            if dergi.kayit_no == guncellenecek_no:
                yeni_baslik = input("Yeni başlık (değişmeyecekse boş bırakın): ")
                yeni_donem = input("Yeni yayın dönemi (değişmeyecekse boş bırakın): ")
                yeni_sayi = input("Yeni sayı numarası (değişmeyecekse boş bırakın): ")

                if yeni_baslik != "":
                    dergi.baslik = yeni_baslik
                if yeni_donem != "":
                    dergi.yayin_donemi = yeni_donem
                if yeni_sayi != "":
                    dergi.sayi_no = yeni_sayi

                print("Dergi güncellendi.")
                return
        print("Kayıt bulunamadı.")

    def listele(self):
        if self.dergi_sayisi() == 0:
            print("Kayıt bulunamadı.")
        else:
            print("\n--- DERGİLER LİSTESİ ---")
            for dergi in self._dergiler:
                print(dergi)


# --- MENÜ SINIFI ---
class Menu:
    def __init__(self):
        self._kitap_islem = KitapIslem()
        self._dergi_islem = DergiIslem()

    def goster(self):
        print("\n--- KÜTÜPHANE YÖNETİM SİSTEMİ ---")
        print("1. Kitap Ekle")
        print("2. Kitap Sil")
        print("3. Kitap Güncelle")
        print("4. Kitapları Listele")
        print("5. Dergi Ekle")
        print("6. Dergi Sil")
        print("7. Dergi Güncelle")
        print("8. Dergileri Listele")
        print("9. Çıkış")

    def calistir(self):
        while True:
            self.goster()
            secim = input("Yapmak istediğiniz işlemi seçin (1-9): ")

            if secim == "1":
                self._kitap_islem.ekle()
            elif secim == "2":
                self._kitap_islem.sil()
            elif secim == "3":
                self._kitap_islem.guncelle()
            elif secim == "4":
                self._kitap_islem.listele()
            elif secim == "5":
                self._dergi_islem.ekle()
            elif secim == "6":
                self._dergi_islem.sil()
            elif secim == "7":
                self._dergi_islem.guncelle()
            elif secim == "8":
                self._dergi_islem.listele()
            elif secim == "9":
                print("Sistemden çıkılıyor. İyi günler!")
                break
            else:
                print("Hatalı seçim yaptınız. Lütfen 1-9 arası bir rakam girin.")

if __name__ == "__main__":
    sistem = Menu()
    sistem.calistir()
```
